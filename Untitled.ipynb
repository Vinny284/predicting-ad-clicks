# Removing outliers from the data

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

df = pd.read_csv('data2.csv')
```

The dataset has 2 classes, click and no click represented by 1 and 0 respectively. The boxplots how that the data contains some outliers.

```python
c = df.loc[df['Click'] == 1]
nc = df.loc[df['Click'] == 0]

websites = df.columns[1:]

for val in websites:
    sns.boxplot(data=df, x='Click', y=val)
    plt.show()
```

The following code sets all points in a category greater than (Q3 + IQR * 1.5) equal to this value. I tried to remove all rows with outliers completely but the dataset became too small and the model did not perform as well.   

```python
for val in websites:    
    q3 = c[val].quantile(0.75)
    iqr = q3-c[val].quantile(0.25)
    thres = c[val].quantile(0.75)+1.5*iqr
    
    col = np.array(c[val])
    for i, s in enumerate(c[val]):
        if s > thres:
            col[i] = thres
    c[val] = col
    
    
for val in websites:
    q3 = nc[val].quantile(0.75)
    iqr = q3-nc[val].quantile(0.25)
    thres = nc[val].quantile(0.75)+1.5*iqr
    
    col = np.array(nc[val])
    for i, s in enumerate(nc[val]):
        if s > thres:
            col[i] = thres
    nc[val] = col

df = pd.concat([c,nc])
```

The new boxplots contain no outliers, which should improve the performance of the model. Some categories like Documents and Social Media Measurment now have 0 varience and are removed. 

```python
indexes = [df.columns[i] for i, val in enumerate(df.var() == 0) if val]
df = df.drop(columns=indexes)

#df.to_csv('data3.csv', index=False)
```
